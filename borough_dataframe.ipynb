```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/Boroughs.csv"

# Read Purchasing File and store into Pandas data frame
boroughs = pd.read_csv(file_to_load)

# set the index to ZIP Code to locate later on
boroughs.set_index("ZIP Code",inplace=True, drop =False)

# output
boroughs.head(10)
```

```python
# set the lists to blank before appending
zip_list = []
neighborhood_list = []
borough_list = []
city_list = []

# loop through the cells in the data frame to pull out the individual zips
for x in boroughs["ZIP Code"]:
    
    # split out the individual zip codes
    zips = x.split(r", ")
    
    # loop through the individual zips that we just pulled out to expand df on ZIP Codes
    for item in zips:
        # appends to lists per category
        zip_list.append(item)
        neighborhood_list.append(boroughs.loc[x,"Neighborhood"])
        borough_list.append(boroughs.loc[x,"Borough"])
        city_list.append(boroughs.loc[x,"City"])

# create a new dataframe where every ZIP Code has it's own row     
NYC_zips = pd.DataFrame(
                        {"City":city_list,
                         "Borough":borough_list,
                         "Neighborhood":neighborhood_list,
                         "ZIP Code":zip_list
                        })
# output
NYC_zips
```

```python
Bronx = []
Brook = []
Stat = []
Queens = []
Man = []


count = 0
for item in NYC_zips["ZIP Code"]:
    bur = NYC_zips.iloc[count,1]
    if bur == "Bronx":
        Bronx.append(item)
    if bur == "Brooklyn":
        Brook.append(item)
    if bur == "Queens":
        Queens.append(item)
    if bur == "Manhattan":
        Man.append(item)
    if bur == "Staten Island":
        Stat.append(item)
    count +=1
    

#lotto_by_burough



mydict = [{'Bronx': len(Bronx), 'Brooklyn': len(Brook), 'Queens': len(Queens),'Manhattan': len(Man),'Staten Island': len(Stat),}]

mydict

lotto_by_burough=pd.DataFrame(mydict)

lotto_by_burough
```
